### **4. Top Student**
**Problem Description**

You are given a list of student records, where each record is represented as a dictionary containing the student's name, a list of test scores, and the weight of each test. Complete the Python function to to find the student with the highest weighted average score.

The weighted average is obtained by multiplying the scores of each test by their respective weights, summing them and dividing it by the total weight.

Note: It is guranteed that no two students will have the same weighted average score.

**Input Format**

A list of student records in the form of dictionaries

**Output Format**

Name of student with highest weighted average

**Example Input**
```Input 1:
[
    {"name": "John", "scores": [90, 85, 88], "weights": [0.3, 0.4, 0.3]},
    {"name": "Alice", "scores": [95, 92, 89], "weights": [0.25, 0.35, 0.4]},
    {"name": "Bob", "scores": [88, 90, 85], "weights": [0.35, 0.3, 0.35]}
]
```

Example Output

```Output 1:
Alice
```

**Example Explanation**

```
Explanation 1:
John => Weighted Average: (90 * 0.3 + 85 * 0.4 + 88 * 0.3) / (0.3 + 0.4 + 0.3) = 87.4
Alice => Weighted Average: (95 * 0.25 + 92 * 0.35 + 89 * 0.4) / (0.25 + 0.35 + 0.4) = 91.55
Bob => Weighted Average: (88 * 0.35 + 90 * 0.3 + 85 * 0.35) / (0.35 + 0.3 + 0.35) = 87.5
```

Therefore, the highest weighted average is achieved by Alice with a score of `91.55`.


````python
records = [
    {"name": "John", "scores": [90, 85, 88], "weights": [0.3, 0.4, 0.3]},
    {"name": "Alice", "scores": [95, 92, 89], "weights": [0.25, 0.35, 0.4]},
    {"name": "Bob", "scores": [88, 90, 85], "weights": [0.35, 0.3, 0.35]}
]
max_wt_score = 0
student = ''
for i in records:
  scores = i['scores']
  wts = i['weights']

  deno = sum(wts)
  wt_sum = 0
  for x in range(len(scores)):
    wt_sum += scores[x]*wts[x]

  wt_score = wt_sum/deno
  if wt_score > max_wt_score:
    max_wt_score = wt_score
    student = i['name']

print(student)
````

````python
# Alternatively
scores = [np.array(scores['scores']) for scores in records]
wt = [np.array(wt['weights']) for wt in records]

wt_scores = [np.sum(scores[i]*wt[i])/np.sum(wt[i]) for i in range(len(scores))]
best_student = [name['name'] for name in records][np.argmax(wt_scores)]
best_student
````

````python
'''
  -- Q1. **Food Delivery II**
  /*
**Problem Statement:**

If the customer's preferred delivery date is the same as the order date, then the order is called immediate; otherwise, it is called scheduled.
The first order of a customer is the order with the earliest order date that the customer made. It is guaranteed that a customer has precisely one first order. Write a query to find the percentage of immediate orders in the first orders of all customers, rounded to 2 decimal places.
*/

-- Solution
WITH CTE as
(SELECT
  *,
  case when _order_date_ = _customer_pref_delivery_date then 'immediate' else 'scheduled' end as delivery_type,
  dense_rank() over(partition by _customer_id__ order by _order_date_) as dk
FROM
  `interview_prep.delivery`)

SELECT ROUND(SUM(CASE WHEN delivery_type = 'immediate' THEN 1 end)*100/ COUNT(*),2) as immediate_percentage
FROM CTE
WHERE dk =1;

  -- Q2. Team Scores in Football Tournament
  /*
*Problem Statement:*

You would like to compute the scores of all teams after all matches. Points are awarded as follows:

- A team receives three points if they win a match (i.e., Scored more goals than the opponent team).
- A team receives one point if they draw a match (i.e., Scored the same number of goals as the opponent team).
- Ateam receives no points if they lose a match (i.e., Scored fewer goals than the opponent team).

Write a query that selects the `team_id` , `team_name` and `num_points` of each team in the tournament after all described matches.

Note: Return the result table ordered by `team_id` in increasing order.
*/
  -- Solution

WITH host as
(SELECT
  host_team,
  CASE
    WHEN m.host_goals > m.guest_goals THEN 3 WHEN m.host_goals = m.guest_goals THEN 1 else 0 end as host_points
FROM
  scaler-dsml-sql-377809.interview_prep.matches as m),

guest as
(SELECT
  guest_team,
  CASE
    WHEN m.guest_goals > m.host_goals THEN 3 WHEN m.host_goals = m.guest_goals THEN 1 else 0 end as guest_points
FROM
  scaler-dsml-sql-377809.interview_prep.matches as m),
host_agg as
(SELECT
  host_team,
  sum(host_points) as num_points
FROM
  host
GROUP BY
  1),
guest_agg as
(SELECT
  guest_team,
  sum(guest_points) as num_points
FROM
  guest
GROUP BY
  1)

select
  t.*,
  ifnull(h.num_points + g.num_points,0) as num_points
from `scaler-dsml-sql-377809.interview_prep.teams` t
left join
  host_agg h
on
  t.team_id = h.host_team
left join guest_agg g
on
  h.host_team = g.guest_team
order by
  t.team_id;

  -- Q3 Movie Rating
  /*
**Problem Statement:**
Write a query to:
 - Find the name of the user who has rated the greatest number of movies. In case of a tie, return the lexicographically smaller user name.
 - Find the movie name with the highest average rating in February 2020. In case of a tie, return the lexicographically smaller movie name.

*/
  -- Solution

SELECT name
FROM (SELECT
    mr.user_id, u.name as name,
    COUNT(*) AS count
  FROM
    `interview_prep.movie rating` mr
INNER JOIN
  `interview_prep.users` u
USING
  (user_id)
GROUP BY
  user_id, name
ORDER BY
  3 DESC, 2
LIMIT 1) tbl1

union all

SELECT
  title
FROM (SELECT
    mr.movie_id,
    m.title,
    AVG(mr.rating) AS avg_rating
  FROM
    `interview_prep.movie rating` mr

  INNER JOIN `interview_prep.movies` m using(movie_id)

  WHERE
    EXTRACT(month
    FROM
      created_at) = 2
  GROUP BY
    1,2
  ORDER BY
    3 DESC, 2
  LIMIT
    1) tbl2;


-- Q4 Facebook app analytics
  /*
Problem Statement:
Assume you have an events table on Facebook app analytics. Write a query to calculate the click-through rate (CTR) for the app in 2022 and round the results to 2 decimal places. Definition and note:
- Percentage of click-through rate (CTR) = 100.0 * Number of clicks / Number of impressions.
- To avoid integer division, multiply the CTR by 100.0, not 100.
- Return the result ordered by app_id in ascending order
*/
  -- Solution

SELECT
  app_id,
  sum(case when event_type ='click' then 1 end)*100.0/ sum(case when event_type ='impression' then 1 end) as ctr
FROM
(SELECT *, EXTRACT(YEAR FROM timestamp) as year
FROM `scaler-dsml-sql-377809.interview_prep.event`) tbl
WHERE (year = 2022)
GROUP BY 1;


  -- Q5. Credit cards issued
  /*
Problem Statement:
Your team at JPMorgan Chase is preparing to launch a new credit card, and to gain some insights, you're analyzing how many credit cards were issued each month.
Write a query that outputs the name of each credit card and the difference in the number of issued cards between the month with the highest issuance cards and the lowest issuance. Arrange the results based on the largest disparity, In case of a tie arrange based on the card name lexicographically.
*/
  -- Soluiton

WITH cc_ranked as
(SELECT
  *,
  dense_rank() over(partition by card_name order by issue_amount desc) as dk_desc,
  dense_rank() over(partition by card_name order by issue_amount ) as dk_asc
FROM
  `scaler-dsml-sql-377809.interview_prep.card_issued`),

c1 as
(SELECT card_name, issue_amount
FROM cc_ranked
WHERE dk_desc = 1),
c2 as
(SELECT card_name, issue_amount
FROM cc_ranked
WHERE dk_asc = 1)

SELECT c1.card_name, c1.issue_amount - c2.issue_amount as difference
FROM c1
LEFT JOIN c2
ON c1.card_name=c2.card_name
ORDER BY 2 DESC;
'''
````
